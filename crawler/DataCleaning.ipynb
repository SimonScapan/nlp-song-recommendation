```python
# aggregate Genres
genres = ['Rock', 'Jazz','Electro','Pop','Sonstige','Alternative/Indie','Psychadelic','Boogie','Metal','Love','Classic','Christian','Hip-Hop','Folk','Latin','Schlager','Rap','Reggae']

def genreaggregator(data, genre_column):

    for index, row in data.iterrows():
        for genre in genres:

            genre_db = get_subgenres(genre)

            if any(subgenre in genre_db for subgenre in data.loc[index,genre_column]): # here is where the magic happen
                data.loc[index,genre_column] = genre
            else:
                data.loc[index,genre_column] = "No genre found"
```

```python
genreaggregator(data,'Song_genre')
```
