```python
def genreaggregator(data, genre_column):
    '''
    Input: DataFrame with Music data and column name with Genre information
    Ouput: aggregated Genre information
    
    Aggregation of subngenres depending on matching list defined in bucket above
    '''
    
    # iterate over rows in DataFrame
    for index, row in data.iterrows():
        song_genres=[]
        
        for song_genre in row.Genre:
            # iterate over genres in maching list above
            for genre in genres:            
               if song_genre in genres[genre]:
                    # append all aggregated genres into a list
                    song_genres.append(genre)

        # error handling if no genres were found (empty list)
        if len(song_genres) > 0:        
            data.loc[index,genre_column] = song_genres
        else:
            data.loc[index,genre_column] = "No genre found"
            
    return data
```

```python
def fill_empty_genre(data, genre_column):
    '''
    Input: DataFrame with Music data and column name with Genre information
    Ouput: aggregated Genre information
    
    If we have no Genre found for a specific Song of a Artist,
    we define the overall genre of the Artist to the Song with no Genre
    '''
    
    # iterate over rows in DataFrame
    for index, row in data.iterrows():
        if row.Song_genre == "No genre found":
            for index2, songs_from_artist in data[data['Artist'] == row.Artist].iterrows():
                if songs_from_artist.Song_genre != "No genre found":
                    data.loc[index,genre_column] = songs_from_artist.Song_genre
        else:
            pass
        
    return data
```

```python
# import raw data from crawler
data = pd.read_pickle('music_data.pkl')

# aggregate Genre information
data = genreaggregator(data,'Genre')

# define genre if no genre is given
data = fill_empty_genre(data,'Genre')

# save data into Pickle file
data.to_pickle("music_data_aggregated_genres.pkl")
```
