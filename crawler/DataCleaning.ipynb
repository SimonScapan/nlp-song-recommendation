```python
def genreaggregator(data, genre_column):

    for index, row in data.iterrows():
        song_genres=[]
        #print(row)
        
        for song_genre in row.Song_genre:
            for genre in genres:            
               if song_genre in genres[genre]:
                    song_genres.append(genre)
        #print(row.Song_genre)
        #print(song_genres)
        if len(song_genres) > 0:        
            data.loc[index,genre_column] = song_genres
        else:
            data.loc[index,genre_column] = "No genre found"
        
```

```python
def fill_empty_genre(data, genre_column):
    for index, row in data.iterrows():
        if row.Song_genre == "No genre found":
            for index2, songs_from_artist in data[data['Artist'] == row.Artist].iterrows():
                #print(songs_from_artist)
                if songs_from_artist.Song_genre != "No genre found":
                    data.loc[index,genre_column] = songs_from_artist.Song_genre
            # break
        else:
            pass
```
